#### 시각화

```python
g = plt.plot(data.index,data["합계 출산율"])
```

```python
# 보조축 설정
fig, axes = plt.subplots(figsize=(10,6))
axes.plot(data["출생아 수"], color="lavender")

axes2 = axes.twinx() # x축 공유
axes2.plot(data["합계 출산율"], color="purple")

plt.show()
```

```python
fig, axes = plt.subplots(figsize=(10,6))
axes.set_ylabel("출생아 수 ( 천 명)")
axes.set_ylim(250,700)
axes.set_yticks([300,400,500,600])
axes.bar(data.index, data["출생아 수"], color="lavender")

# bar 숫자 표시
for idx, val in enumerate(data["출생아 수"]):
    axes.text(idx, val+12, val, ha="center")

axes2 = axes.twinx() # x축 공유
axes2.plot(data["합계 출산율"], color="purple", marker="o", ms=12, lw=5, mec="w",mew=3)
axes2.set_ylabel("합계 출산율(가임여성 1명당 명)")
axes2.set_ylim(0,1.5)
axes2.set_yticks([0,1])

# 꺾은선 숫자 보여주기
for idx, val in enumerate(data["합계 출산율"]):
    axes2.text(idx, val+0.06, val, ha="center")

fig.suptitle("출생아 수 합계 출산율", fontsize=15)

plt.show()
```
