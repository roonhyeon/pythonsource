##### 데이터 전처리
    - 수치 데이터 object => int 64로 변경(df.astype(), pd.to_numeric())

```python
import numpy as np

df = pd.DataFrame({
    "name":["홍길동","김수정","이민정"],
    "age":["25",np.nan,"45"]
})
df
```

```python
df.dtypes
```

```python
# astype(): 숫자에 , 가 없어야 함 / NaN 없어야 함
# df["age1"] = df["age"].astype("int64")
```

```python
# , 제거 후 astype()
# replace(",","")
df_man["0~4세"].str.replace(",","").astype("int64")
```

```python
# 모든 컬럼에 ,를 제거하고 데이터 타입 변경 => 함수 작성
df_man = df_man.apply(lambda x:x.str.replace(",","").astype("int64"))
df_women = df_women.apply(lambda x:x.str.replace(",","").astype("int64"))
```

```python
df_man.dtypes
```

- 남자, 여자 컬럼명 통일

```python
df_man.columns
```

```python
df_man.columns = df_women.columns
```

```python
df_man.iloc[0]
```

```python
plt.figure(figsize=(10,7))
plt.barh(y=df_man.columns, width=-df_man.iloc[0] // 10000)
plt.barh(y=df_women.columns, width=df_women.iloc[0] // 10000)
plt.title("2023 6월 기준 대한민국 인구 피라미드")
plt.savefig("202306_인구피라미드.png")
plt.show()
```
