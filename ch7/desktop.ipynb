```python
import pyautogui

# 현재 화면의 스크린 사이즈 가져오기
size = pyautogui.size()
print(size)
```

```python
# 마우스 핸들링
import pyautogui as p

# 마우스 위치 좌표 구하기
position = p.position()
print(position.x, position.y)
```

```python
# moveTo(x좌표, y좌표): 절대좌표
p.moveTo(100,100,0.5)
p.moveTo(200,200,0.5)
p.moveTo(300,300,0.5)
```

```python
# move(), moveRel(): 상대좌표
p.moveTo(300,300,0.25)
p.move(100,100,0.25)
print(p.position())
p.move(200,200,0.25) 
```

```python
p.sleep(3)
print(p.position())
```

```python
p.sleep(1)
p.click(clicks=2)
```

```python
p.doubleClick(487,142)
```

```python
p.rightClick(500,300)
```

```python
# mouseDown() / mouseUp()
p.sleep(3)
p.moveTo(600,600)
p.mouseDown()
p.moveTo(700,700)
p.mouseUp()
```

```python
# 전체 화면 스크린샷
p.screenshot("screen.png")
```

```python
# 특정 위치의 스크린샷
# region(x축 상단, y축 상단, width, height)
p.screenshot("region.png",region=(0,0,300,400))
```

```python
# locateOnScreen: 이미지의 화면상 좌표 구하기
screen_locate = p.locateOnScreen("screen.png")
print(screen_locate)
```

```python
p.screenshot("file.png",region=(0,0,300,400))
```

```python
file_locate = p.locateOnScreen("file.png", confidence=0.9)
print(file_locate)
p.click(file_locate)
```

```python
p.screenshot("checkbox.png",region=(0,0,300,400))
```

```python
checkbox = p.locateOnScreen("checkbox.png", confidence=0.9)
print(checkbox)
```

```python
import pyperclip

# 메모장 가져오기
w = p.getWindowsWithTitle("제목 없음")[0]

# 메모장 활성화
w.activate()

# 클립보드 복사
pyperclip.copy("파이썬")

# 클립보드 내용 붙여넣기
p.hotkey("ctrl","v")
```

```python
def word_write(text):
    pyperclip.copy(text)
    p.hotkey("ctrl","v")
```

```python
word_write('abcdefghijklmnopqrstuvwsyz')
word_write('가나다라마바사아자차카타파하')
```
