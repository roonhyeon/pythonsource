```python
# 새 탭 열기와 같은 작업은 스크립트로 실행: execute_script(스크립트 코드 작성)
from selenium.webdriver.common.by import By # 파싱 기준
from selenium.webdriver.common.keys import Keys # 키보드 키 지정
import time

driver = set_chrome_driver()
driver.get("http://www.naver.com")

driver.execute_script("window.open('https://www.daum.net')") # 새 탭 열기

# 브라우저 열린 탭 정보 가져오기
tabs = driver.window_handles

# print(tabs)
driver.switch_to.window(tabs[1])

element = driver.find_element(By.NAME,"q")

# 검색어 입력
element.send_keys("아이폰")

# 엔터
element.send_keys(Keys.ENTER)

time.sleep(2)

# 네이버로 돌아오기
driver.switch_to.window(tabs[0])
```

```python
# 네이버 쇼핑 / 검색어 '마우스'
driver = set_chrome_driver()
driver.get("https://shopping.naver.com/home")

element = driver.find_element(By.CLASS_NAME, "_searchInput_search_text_3CUDs")

# 검색어 입력
element.send_keys("마우스")

# 텍스트에서 엔터
element.send_keys(Keys.ENTER)

time.sleep(2)

# window.scrollTo(0,1080)
# driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")

# 외부 스크롤
# 현재 문서 높이
prev_height = driver.execute_script("return document.body.scrollHeight")

# 스크롤 이동 시간 간격(2초 마다 내려줘)
interval = 2

while True:
    # 스크롤 이동
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")

    # 페이지 로딩 대기
    time.sleep(interval)

    # 스크롤이 진행된 후 현재 문서 높이
    curr_height = driver.execute_script("return document.body.scrollHeight")

    if curr_height == prev_height:
        break
    
    prev_height = curr_height

driver.execute_script("window.scrollTo(0,0)")
```

```python
# 구글 이미지 다운로드
from urllib.request import urlretrieve


driver = set_chrome_driver()
driver.get("https://www.google.com/imghp?hl=ko&tab=ri&ogbl")

# 검색 창 가져오기
element = driver.find_element(By.NAME, "q")
element.send_keys("python")
element.send_keys(Keys.ENTER)

# 현재 문서 높이
prev_height = driver.execute_script("return document.body.scrollHeight")

# 스크롤 이동 시간 간격
interval = 2

while True:

    # 스크롤 이동
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")

    # 페이지 로딩 대기
    time.sleep(interval)

    # 스크롤이 진행된 후 현재 문서 높이
    curr_height = driver.execute_script("return document.body.scrollHeight")

    if curr_height == prev_height:
        try:
            # 결과 더보기 클릭
            driver.find_element(By.CLASS_NAME, "LZ4I").click()
        except:
            break

    prev_height = curr_height

# 이미지 저장

# 전체 이미지 가져오기
images = driver.find_elements(By.CSS_SELECTOR, "div.bRMDJf.islir > img")
count = 1
for img in images:

    try:
        # 작은 이미지 클릭
        img.click()

        time.sleep(2)

        # 큰 이미지 경로 찾기
        imgUrl = driver.find_element(By.XPATH, "//*[@id='Sva75c']/div[2]/div[2]/div[2]/div[2]/c-wiz/div/div/div/div[3]/div[1]/a/img").get_attribute("src")
        # print(imgUrl)

        # urlretrieve() 저장
        urlretrieve(imgUrl, "./download/"+str(count)+".jpg")
        count += 1
    except:
        pass
```

```python
# driver.implicity_wait(3): 3초 안에 웹페이지가 로드되면, 바로 넘어가버림
# time.sleep(3): 무조건 3초
import requests
from fake_useragent import UserAgent

from io import BytesIO
from openpyxl import Workbook


# 엑셀 작업
wb = Workbook()
ws = wb.active

ws.column_dimensions["A"].width = 60
ws.column_dimensions["B"].width = 20

# 제목행 추가
ws.append(["제품명", "가격", "이미지"])

from bs4 import BeautifulSoup

# 다나와 노트북 선택
url = "https://prod.danawa.com/list/?cate=112758&15main_11_02"
driver = set_chrome_driver()
driver.get(url)

# 제조사 22개 클릭(WebDriverWait() 사용) => apple 클릭(WebDriverWait() 사용)
try:
    WebDriverWait(driver,3).until(EC.presence_of_element_located((By.XPATH, '//*[@id="dlMaker_simple"]/dd/div[2]/button[1]'))).click()

    WebDriverWait(driver,3).until(EC.presence_of_element_located((By.XPATH, '//*[@id="selectMaker_simple_priceCompare_A"]/li[16]/label'))).click()

    # 애플 제품 로딩
    # driver.implicitly_wait(3)
    time.sleep(3)

    # 페이지 나누기 변수
    cur_page, target_crawl_num = 1,4
    # 전체 상품 번호
    idx = 1
    # 엑셀 행 번호
    ins_cnt = 2

    while cur_page <= target_crawl_num:
        soup = BeautifulSoup(driver.page_source, "html.parser")

        # 상품목록 리스트 모두 가져오기 li
        prod_list = soup.select("div.main_prodlist.main_prodlist_list > ul > li:not(.prod_ad_item):not(.product-pot)")
        # print(prod_list)

        for prod in prod_list:
            # 번호, 제품명, 가격, 이미지주소
            prod_name = prod.select_one(".prod_name a").get_text().strip()
            prod_price = prod.select_one("p.price_sect strong").get_text().strip()
            img = prod.select_one(".thumb_image img")

            if img.get("data-original"):
                prod_img = img.get("data-original")
            else:
                prod_img = img.get("src")

            if "http:" not in prod_img:
                prod_img = "http:"+prod_img

            print(idx, prod_name, prod_price, prod_img)
            
            # 이미지 요청
            res = requests.get(prod_img, headers = {"user-agent": UserAgent().chrome})
            save_img = BytesIO(res.content)

            # 상품 하나에 대한 정보 저장
            ws.append([prod_name, prod_price])

            # 엑셀에 이미지 저장
            image = Image(save_img)
            image.width = 30
            image.height = 20
            ws.add_image(image, "C"+str(ins_cnt))

            idx += 1
            ins_cnt += 1

        # 현재 페이지 번호 변경
        cur_page += 1

        # 다음 페이지 번호 클릭
        WebDriverWait(driver,2).until(
            EC.presence_of_element_located(
                (By.CSS_SELECTOR, "div.number_wrap > a:nth-child({})".format(cur_page))
            )
        ).click()

        # 기존의 soup 삭제
        del soup
        
        time.sleep(3)
    
except TimeoutException as e:
    print(e)

wb.save("./apple.xlsx")
wb.close()
```
