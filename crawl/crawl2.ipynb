```python
# get 방식

import requests

# 세션을 이용한 요청 가능
s = requests.Session()

r = s.get("https://naver.com")

print(r.status_code)
print(r.ok)

print(r.text)

# 세션 종료
s.close()
```

```python
# get / post / put / delete 함수 제공함

import requests

s = requests.Session()

# r = s.get("https://httpbin.org/get")
# post + data
data = {
    "name":"hong"
}
r = s.post("https://httpbin.org/post",data=data)

print(r.text)

s.close()
```

```python
import requests

s = requests.Session()

# put + data
data = {
    "name":"hong"
}
r = s.put("https://httpbin.org/put",data=data)

print(r.text)

s.close()
```

```python
import requests

s = requests.Session()

param = {"name":"hong", "age":"14"}

r = s.get("https://httpbin.org/get",params=param)

print(r.text)

s.close()
```

```python
import requests

r = requests.get("https://jsonplaceholder.typicode.com/todos/1")
# print(r.text)
print(r.headers)
print(r.encoding)
print(r.json())
print(r.json().keys())
print(r.json().values())
```

```python
r = requests.get("https://jsonplaceholder.typicode.com/users")

for row in r.json():
    for k, v in row.items():
        print(k, ":", v)
    print()
```
