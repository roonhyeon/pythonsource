```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
plt.rcParams["font.family"]="Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
import warnings
warnings.simplefilter("ignore")
```

```python
df = pd.DataFrame(np.random.randn(1000,3), columns=list('ABC'))
df = df.cumsum()
df
```

```python
# 범례가 포함된 선 차트
g = df.plot()
```

```python
g = df.plot(legend=False)
```

```python
# g = df["B"].plot(legend=False)
g = df.plot(y="B")
```

```python
# 보조축 + 레이블 설정
ax = df.plot(secondary_y=["A","B"], mark_right=False)
ax.set_ylabel("C_Scale")
ax.right_ax.set_ylabel("AB_Scale")
plt.show()
```

```python
# 컬럼별 선 차트 따로 그리기
g = df.plot(subplots=True)
```

```python
np.random.seed(1000)
df = pd.DataFrame(np.random.randn(1000,4), columns=list('ABCD'), index=pd.date_range('1/1/2021',periods=1000))
df
```

```python
# 2021-01-06 날짜로 바 차트 작성
# df["B"].plot()
# iloc: 숫자, loc: 레이블명
df.iloc[5].plot(kind="bar")
```

```python
df2 = pd.DataFrame(np.random.rand(10,4), columns=list('ABCD'))
df2. plot.bar()
```

```python
# 누적 막대
df2.plot.bar(stacked=True)
```

```python
df2.plot.box()
```

```python
df2 = pd.DataFrame(np.random.rand(50,4), columns=list('abcd'))
```

```python
g = df2.plot.scatter(x="a",y="b")
```

```python
# s: 점 크기 설정
g = df2.plot.scatter(x="a",y="b",s=50)
```

```python
ax = df2.plot.scatter(x="a",y="b",color="DarkBlue",label="Group 1")
df2.plot.scatter(x="c",y="d",color="DarkGreen",label="Group 2",ax=ax)
```

```python
series = pd.Series(np.random.rand(4)*3, index=list('abcd'), name="series")
series
```

```python
# matplotlib 차트 크기변경 or pandas 이용
# plt.figure(figsize=(3,3))
series.plot.pie(figsize=(3,3),autopct="%.1f%%")
```
