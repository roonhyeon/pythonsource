```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.rcParams["font.family"]="Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
df = pd.read_csv("./student.csv")
df
```

### subplots()

```python
fig, axes = plt.subplots(2,2,figsize=(15,10))
fig.suptitle("여러 개의 그래프")

# 첫번째 그래프
axes[0,0].bar(df["이름"],df["국어"])
axes[0,0].set_title("학생별 국어 점수")
axes[0,0].legend(["국어 점수"])
axes[0,0].set(xlabel="이름",ylabel="점수")
axes[0,0].set_facecolor("lightyellow")
axes[0,0].grid(linestyle="--", linewidth=0.5)

# 두번째 그래프
axes[0,1].plot(df["이름"],df["영어"])
axes[0,1].plot(df["이름"],df["수학"])
axes[0,1].set_title("학생별 영어/수학 점수")
axes[0,1].legend(["영어 점수","수학 점수"])
axes[0,1].set(xlabel="이름",ylabel="점수")
axes[0,1].set_facecolor("lightyellow")
axes[0,1].grid(linestyle="--", linewidth=0.5)

# 세번째 그래프
axes[1,0].barh(df["이름"],df["키"])
axes[1,0].set_title("학생별 키")
axes[1,0].set(xlabel="이름",ylabel="키")

# 네번째 그래프
axes[1,1].plot(df["이름"],df["사회"],color="green",alpha=0.5)
```
