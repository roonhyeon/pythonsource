### [퀴즈1]
#### [문제] 다음은 대한민국 영화 중에서 관객 수가 가장 많은 상위 10 개의 데이터입니다.
#### 주어진 코드를 이용하여 퀴즈를 풀어보세요

```python
import pandas as pd
```

```python
dict1 = {"영화":["명량","극한직업","신과함께 - 죄와 벌","국제시장","베테랑","괴물","도둑들","7번방의 선물","암살","광해, 왕이 된 남자",], "개봉 연도":[2014,2019,2017,2014,2015,2006,2012,2013,2015,2012], "관객 수":[1761,1626,1441,1425,1341,1301,1298,1281,1270,1231], "평점":[8.88,9.20,8.73,9.16,9.24,8.62,7.64,8.83,9.10,9.25]}
movie_df = pd.DataFrame(dict1)
movie_df
```

#### 1) 전체 데이터 중에서 '영화' 정보만 출력하기

```python
movie_df.loc[:,["영화"]]
```

#### 2) 전체 데이터 중에서 '영화','평점' 정보 출력하기

```python
movie_df.loc[:,["영화","평점"]]
```

#### 3) 2015년 이후에 개봉한 영화 데이터 중에서 '영화','개봉연도' 정보만 출력하기

```python
movie_df[movie_df["개봉 연도"] >= 2015][["영화","개봉 연도"]]
```

#### 4) 주어진 계산식을 참고하여 '추천점수' 컬럼 추가
> 추천점수 = (관객수 * 평점) // 100

```python
movie_df["추천점수"]=(movie_df["관객 수"] * movie_df["평점"]) // 100
movie_df
```

#### 5) 전체 데이터를 '개봉 연도' 기준 내림차순으로 출력하기

```python
movie_df.sort_values("개봉 연도",ascending=False,ignore_index=True)
```

### [퀴즈5]

#### 1) data/covid/04-01-2020.csv 파일 dataframe 생성 (데이터프레임 변수 이름은 covid_df 로  지정)

```python
covid_df = pd.read_csv("./data/covid/04-01-2020.csv", encoding="cp949")
covid_df.head(3)
```

#### 2) covid_df 에서 중복된 Country_Region 각각에 대해 Confirmed 를 모두 더한 값을 컬럼으로 갖는 데이터프레임을 피봇테이블 작성

```python
pd.pivot_table(covid_df, index="Country_Region", values="Confirmed", aggfunc="sum")
```

#### 3) covid_df 에서 중복된 Country_Region 각각에 대해 Confirmed 를 모두 더한 값을 컬럼으로 갖는 데이터프레임을 피봇테이블로  만들고, 총 값을 'Total' 을 인덱스 값으로 하는 record 로 추가하세요
- 결측치는 0으로 처리

```python
pd.pivot_table(covid_df, index="Country_Region", values="Confirmed", aggfunc="sum", fill_value=0, margins=True, margins_name="Total")
```
